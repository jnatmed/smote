```python
from pandas import DataFrame
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('credit_dataset.csv')
# print(df.head())

x_values = df['TARGET'].unique()
y_values = df['TARGET'].value_counts().tolist()
plt.bar(x_values, y_values)
plt.title('Grafico de barras de la variable TARGET')
ax = plt.subplot()                  #Ejes
ax.set_xticks(x_values)             #Eje x
ax.set_xticklabels(x_values)        #Etiquetas del eje x
ax.set_xlabel('Fraude tarjeta de Credito (0 = normal, 1 = fraude)')  #Nombre del eje x
ax.set_ylabel('Numero de Transacciones')  #Nombre del eje y
plt.show()
plt.close('all')
```

- Como se puede observar las clases estan desbalanceadas.
